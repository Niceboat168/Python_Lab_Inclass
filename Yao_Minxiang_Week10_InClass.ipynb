```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("[UCI] AAAI-14 Accepted Papers - Papers.csv",index_col=0)
```

```python
 df['authorsnumber']=0
```

```python
i = 0
while i < 398:   
    df['authorsnumber'][i] = df['authors'][i].count(',')+df['authors'][i].count('and')+1
    i = i+1
```

```python
df['groupsnumber']=0
```

```python
i = 0
while i < 398:   
    df['groupsnumber'][i] = str(df['groups'][i]).count(')')
    i = i+1
```

```python
df['keysnumber']=0
```

```python
i = 0
while i < 398:   
    df['keysnumber'][i] = str(df['keywords'][i]).count('\n')+1
    i = i+1
```

```python
df['topicsnumber']=0
```

```python
i = 0
while i < 398:   
    df['topicsnumber'][i] = str(df['topics'][i]).count('\n')
    i = i+1
```

```python
df
```

```python
sns.lmplot(x='authorsnumber',y='groupsnumber',data=df,fit_reg=False,size=10)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
df=df.drop('abstract',axis=1)
df= df.drop('keywords',axis=1)
df= df.drop('topics',axis=1)
df=df.drop('authors',axis=1)
df=df.drop('groups',axis=1)
```

```python
df
```

```python
kmeans.fit(df)
```

```python
kmeans.cluster_centers_
```

```python
df['Cluster']=0
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print("\n")
print (classification_report(df['Cluster'],kmeans.labels_))
```
