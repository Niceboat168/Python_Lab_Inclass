```python
y = mpg['mpg']
```

```python
x = mpg[['weight', 'acceleration', 'horsepower','displacement']]
```

** Use model_selection.train_test_split from sklearn to split the data into training and testing sets. Set test_size=25/75 and random_state=2018**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=25/75, random_state=2018)
```

```python
from sklearn.linear_model import LinearRegression
```

**Create an instance of a LinearRegression() model named lm.**

```python
lm = LinearRegression()
```

** Train/fit lm on the training data.**

```python
lm.fit(x_train,y_train)
```

**Print out the coefficients of the model**

```python
print('Coefficients: \n', lm.coef_)
```

```python
predictions = lm.predict( x_test)
```

** Create a scatterplot of the real test values versus the predicted values. **

```python
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

## Evaluating the Model

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

## Residuals

```python
sns.distplot((y_test-predictions),bins=50);
```

## Conclusion

```python
coeffecients = pd.DataFrame(lm.coef_,x.columns)
coeffecients.columns = ['Coeffecient']
coeffecients
```

** How can you interpret these coefficients? **

Interpreting the coefficients:

- Holding all other features fixed, a 1 unit increase in **weight** is associated with a **decrease of 0.006410**.
- Holding all other features fixed, a 1 unit increase in **acceleration** is associated with a **decrease of 0.013672 mpg**.
- Holding all other features fixed, a 1 unit increase in **horsepower** is associated with a **decrease of 0.050176 mpg**.
- Holding all other features fixed, a 1 unit increase in **displacement** is associated with an **increase of 0.005543 mpg**.
